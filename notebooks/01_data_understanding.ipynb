```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os, sys

sys.path.append(os.path.join('..', 'src'))
from preprocessing import generate_dataset, FEATURES, CLASS_PARAMS

sns.set_theme(style='whitegrid')
pd.set_option('display.float_format', '{:.3f}'.format)
print('Libraries loaded ✓')
```

## 1. Generate / Load Dataset

```python
# Run preprocessing to generate dataset if it doesn't exist
raw_path = os.path.join('..', 'dataset', 'raw_data', 'soil_data.csv')

if not os.path.exists(raw_path):
    print('Generating dataset...')
    df = generate_dataset()
    os.makedirs(os.path.dirname(raw_path), exist_ok=True)
    df.to_csv(raw_path, index=False)
    print(f'Saved to {raw_path}')
else:
    df = pd.read_csv(raw_path)
    print(f'Loaded existing dataset from {raw_path}')

print(f'Shape: {df.shape}')
```

## 3. Data Quality Check

```python
# Missing values
nulls = df.isnull().sum()
print('Missing Values per Column:')
print(nulls)
print(f'\nTotal missing: {nulls.sum()}')
```

```python
# Duplicates
dupes = df.duplicated().sum()
print(f'Duplicate rows: {dupes}')
```

## 4. Descriptive Statistics

```python
df[FEATURES].describe().round(2)
```

```python
print('Skewness (absolute > 1 suggests high skew):')
print(df[FEATURES].skew().round(3))
print('\nKurtosis:')
print(df[FEATURES].kurt().round(3))
```

## 5. Target Class Distribution

```python
class_counts = df['Fertility_Class'].value_counts()
print('Class Distribution:')
print(class_counts)
print(f'\nBalance ratio (max/min): {class_counts.max()/class_counts.min():.2f}')
```

```python
CLASS_ORDER = ['High Fertility', 'Medium Fertility', 'Low Fertility', 'Poor Fertility']
PALETTE = {'High Fertility':'#3B6D11', 'Medium Fertility':'#97C459', 'Low Fertility':'#EF9F27', 'Poor Fertility':'#E24B4A'}

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Bar
counts_ordered = df['Fertility_Class'].value_counts()[CLASS_ORDER]
axes[0].bar(CLASS_ORDER, counts_ordered, color=[PALETTE[c] for c in CLASS_ORDER], edgecolor='none')
axes[0].set_title('Class Count (Bar)', fontweight='bold')
axes[0].set_xticklabels(['High', 'Medium', 'Low', 'Poor'])
axes[0].set_ylabel('Count')
axes[0].grid(axis='y', linewidth=0.5)

# Pie
axes[1].pie(counts_ordered, labels=CLASS_ORDER,
            colors=[PALETTE[c] for c in CLASS_ORDER],
            autopct='%1.1f%%', startangle=140,
            wedgeprops={'edgecolor':'white','linewidth':1.5})
axes[1].set_title('Class Distribution (Pie)', fontweight='bold')

plt.suptitle('Target Class Distribution', fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

## 6. Feature Range per Class

```python
# Mean of each feature per class
summary = df.groupby('Fertility_Class')[FEATURES].mean().round(2)
summary.loc[CLASS_ORDER]
```

```python
# Per-class std
df.groupby('Fertility_Class')[FEATURES].std().round(2).loc[CLASS_ORDER]
```
